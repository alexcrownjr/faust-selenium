# Compare crawls

```python
orig = sqlite3.connect('/home/bird/Dev/mozilla/OpenWPM/crawls/crawl-data-6/crawl-data.sqlite')
new = sqlite3.connect('crawls/crawl-data-6.sqlite') 
```

```python
crawl_requests_orig = pd.read_sql('SELECT * FROM crawl_history;', orig)
crawl_requests_new = pd.read_sql('SELECT * FROM crawl_requests;', new)
crawl_requests = crawl_requests_orig.merge(crawl_requests_new, left_on='arguments', right_on='url', suffixes=('_orig', '_new'))
crawl_requests = crawl_requests[['visit_id_orig', 'visit_id_new', 'url']]
print(len(crawl_requests))
crawl_requests.head()
```

```python
comparison_by_url = {}
for i, row in tqdm(crawl_requests.iterrows()):
    orig_requests = pd.read_sql(f'SELECT id FROM http_requests WHERE visit_id == {row.visit_id_orig};', orig, index_col='id') 
    new_requests = pd.read_sql(f'SELECT id FROM http_requests WHERE visit_id == {row.visit_id_new};', new, index_col='id') 
    orig_javascript = pd.read_sql(f'SELECT id FROM javascript WHERE visit_id == {row.visit_id_orig};', orig, index_col='id') 
    new_javascript = pd.read_sql(f'SELECT id FROM javascript WHERE visit_id == {row.visit_id_new};', new, index_col='id') 
    comparison_by_url[row.url] = {
        'http_request': (len(orig_requests), len(new_requests), len(new_requests) - len(orig_requests)),
        'javascript': (len(orig_javascript), len(new_javascript), len(new_javascript) - len(orig_javascript)),
    }
```

```python
diff = {k: x['http_request'] for k, x in comparison_by_url.items()}
sorted(diff.items(), key=lambda x: x[1][2])[0:20]
```

```python
diff = {k: x['javascript'] for k, x in comparison_by_url.items()}
sorted(diff.items(), key=lambda x: x[1][2])[0:20]
```
